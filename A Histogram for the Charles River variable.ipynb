```python
    "# Imported libraries for data analysis and visualization"
    import pandas as pd 
    import numpy as np
    import seaborn as sns
    import matplotlib.pyplot as plt
    import scipy.stats
    import statsmodels.api as sm
```

```python
"#Loaded the Boston housing dataset into a DataFrame using pandas"
boston_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ST0151EN-SkillsNetwork/labs/boston_housing.csv'
boston_df = pd.read_csv(boston_url)
```

```python
"#The first 10 rows of the Boston housing dataset"
boston_df.head(10)
```

```python
    "# Check the first few rows to understand the data",
    print(boston_df.head())

    "# Create a histogram for 'CHAS' (Charles River variable)\n",
    plt.figure(figsize=(10, 6))
    plt.hist(boston_df['CHAS'], bins=2, edgecolor='black', alpha=0.7)
    plt.title('Histogram of Charles River Variable')
    plt.xlabel('Charles River (1 = Near, 0 = Not Near)')
    plt.ylabel('Frequency')
    plt.xticks([0, 1], ['Not Near', 'Near'])
    plt.grid(True)
    plt.show()
   
  
```
